# NumPy

NumPy is the numerical engine beneath Pandas, scikit-learn, TensorFlow, and PyTorch. Every dataset 
eventually becomes a NumPy array. This chapter builds the mental model you need before touching any 
ML library. 

If previous chapter  was about reading Python, this chapter is about thinking in arrays. Once you can 
comfortably visualize an operation happening across an entire array, not element by element, the 
rest of the ML stack becomes far easier to learn. 

NumPy provides the ndarray-> a fast, fixed-type, multi-dimensional array, plus vectorized math 
operations written in C

Python lists are slow for numeric work because each element is a separate object. NumPy stores data in 
contiguous memory and applies operations to entire arrays at once. 

#AI/ML Use Case 

Every image, audio signal, weight matrix, and gradient in ML is represented as a NumPy array (or a 
library built on the same idea, like a PyTorch tensor

NumPy arrays are homogeneous (single data type) and fixed-size, unlike Python lists ...this is 
exactly why they're fast.

# Broadcasting and Vectorization 

Broadcasting lets NumPy apply operations between arrays of different (but compatible) shapes without 
writing loops. Vectorization means applying an operation to a whole array at once. 

```python
a = np.array([1, 2, 3]) 
a * 2                 # [2, 4, 6]  -- scalar broadcast to every element

m = np.array([[1,2,3],[4,5,6]]) 
v = np.array([10, 20, 30]) 
m + v               # v is broadcast across every row     
# [[11,22,33], 
#  [14,25,36]] 
```

#AI/ML Use Case 

Normalizing every row of a feature matrix by subtracting the column means — done in one broadcasted 
line instead of nested loops

Vectorized NumPy code can be 50-100x faster than an equivalent Python for-loop. Interviewers 
often ask you to vectorize a loop.

# Mathematical Operations & Aggregation

```python
a = np.array([[1,2,3],[4,5,6]]) 
  
a.sum()          # 21 (all elements) 
a.mean()         # 3.5 
a.std()          # standard deviation 
a.max(); a.min() 
a.sum(axis=0)    # column-wise sum -> [5, 7, 9] 
a.sum(axis=1)    # row-wise sum    -> [6, 15] 
```

AI/ML Use Case 

axis=0 vs axis=1 is the single most confused — and most tested — NumPy concept. axis=0 collapses rows 
(operates down each column); axis=1 collapses columns (operates across each row).

# Copy vs View 

slicing a numpy array returns a view(shares memory with the original) by default, not a copy.

Modifying a view changes the original array

```python
a = np.array([1,2,3,4,5])

b = a[0:3]    # view -- shares memory with a
```

```python
b
```

```python
b[0] = 99
```

```python
a         # a changed too
```

```python
# explicit copy -- safe, independent
c = a[0:3].copy()
c
```

# Handling NaN in Numpy

np.nan represents a missing/undefined float value. 

It behaves unusually: it is never equal to anything, 
including itself.

```python
a = np.array([1, np.nan, 3])
a
```

```python
np.isnan(a)   
```

```python
np.nanmean(a)        # mean ignoring nan
```

```python
# to filter out NaN values
a[np.isnan(a) == False]

# another way to write is using predicate logic, negation sign
a[~np.isnan(a)]
```

# Performance Benefits 
NumPy operations run in compiled C code, use contiguous memory blocks, and avoid Python's per
object overhead. In practice, this means: 


• Vectorized operations replace explicit Python loops 

• Memory usage is lower, fixed-type arrays vs. heterogeneous Python objects 

• Operations scale to millions of rows, which is essential for real datasets 

Interview Tips

If asked 'why is NumPy faster than lists', mention: contiguous memory, fixed dtype, and 
vectorized C-level loops — not just 'it's optimized'. 

# Worked Example : Linear Regression by Hard

```python
X = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])
```

```python
x_mean, y_mean = X.mean(), y.mean()
```

```python
# slope (m) and intercept (b) using the least - squares formula
```

```python
m = np.sum((X-x_mean) *( y-y_mean))/ np.sum((X-x_mean)**2)

b = y_mean -m * x_mean

predictions = m*X+b
```

```python
predictions
```

Every single piece of this — mean(), broadcasting (X - x_mean), element-wise multiplication, and sum() 
is something covered in this chapter. This is exactly the math scikit-learn's LinearRegression runs 
internally, just wrapped in a friendlier interface. 

Interview Tip:  Being able to derive linear regression with raw NumPy is a strong interview signal 
,it shows you understand what the library is actually doing, not just how to call .fit().
