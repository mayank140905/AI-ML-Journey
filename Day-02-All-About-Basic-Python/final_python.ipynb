## collections 

```python
from collections import Counter 
Counter(["a","b","a","c","a"]) 
# Counter({'a': 3, 'b': 1, 'c': 1})  -- great for class balance checks 
```

```python
from collections import defaultdict 
groups = defaultdict(list) 
for name, dept in [("Asha","Eng"), ("Rohan","Sales"), ("Mira","Eng")]: 
    groups[dept].append(name) 
# {"Eng": ["Asha","Mira"], "Sales": ["Rohan"]}  -- no KeyError on new keys 
```

Interview Tip:  defaultdict removes the need to check 'if key not in dict' before appending 
,it auto creates a default value (here, an empty list) for any new key.

## Generators
A generator produces values one at a time, on demand, instead of building a full list in memory. You 
rarely write these from scratch in ML, but data loaders (e.g. PyTorch DataLoader) are built on this idea.

```python
def batch_generator(data , batch_size):
    for i in range(0, len(data), batch_size):
        yield data [i:i + batch_size]

for batch in batch_generator(range(10),3):
    print(batch)    
```

#AI/ML Use Case 

Generators allow training on datasets too large to fit in memory by yielding one batch at a time instead 
of loading everything at once

#Interview Tip 

Tip:  The keyword that turns any function into a generator is yield. A function with yield doesn't run until 
you iterate over it.

1) Write a function that takes a list of exam scores and returns how many are passing (score >= 40).

```python
def check_pass(scores):
    return len([s for s in scores if s>=40])
```

2) Given a dictionary of {student: score}, return the name of the student with the highest score. 

```python
def highest_score(scores):
    return max(scores, key = scores.get())

# key = scores.get() --> returns value
```

3. Write a function using try/except that safely converts a string to a float, returning None on failure. 

```python
def safe_float(s):
    try:
        return float(s)
    except ValueError:
        return None
```
