# Exception Handling

A mechanism to catch and respond to runtime errors gracefully instead of letting the entire program 
crash. 

Why do we need it? 

Real pipelines hit messy, unpredictable data.... a missing file, a corrupt row, a divide-by-zero. Exception 
handling lets you respond intelligently (skip, log, default) instead of crashing the whole job.

```python
try:
    value = 10/ divisor
except ZeroDivisionError:
    print("Cannot divide by zero")
except Exception as e:
    print(f"Unexpected error: {e}")
finally:
    print("done")
```

```python
### Example

def safe_divide(a,b):
    try:
        return a/b
    except ZeroDivisionError:
        return None 

print(safe_divide(10,0)) #none, no crash
```

Common Exception Types 

Exception            Typical Trigger 
ValueError           Invalid value, e.g. int("abc") 
TypeError            Wrong type used in an operation 
KeyError             Missing dictionary key 
IndexError           List/array index out of range 
FileNotFoundError    Opening a file that doesn't exist 
ZeroDivisionError    Dividing by zero

AI/ML Use Case 

Wrapping data-loading or API calls so one bad file/row doesn't crash an entire pipeline. 

Interview Tip 

Tip:  finally always runs, whether or not an exception occurred , it's the right place for cleanup like closing a file or connection. 


Common Mistakes 

Mistake:  Using a bare except: that silently swallows every error, hiding real bugs. Always catch specific 
exceptions. 

# Basic Classes and Objects

A blueprint for creating objects with attributes and methods. You rarely write deep OOP in ML, but you 
must read it, every model in scikit-learn/PyTorch is a class. 

```python
class Model:
    def __init__(self, learning_rate):  
        self.learning_rate = learning_rate     # attribute
        self.is_trained = False 

    def train(self):                 #method
        self.is_trained = True
        return "Training Complete"

m = Model(0.01)
print(m.learning_rate)
print(m.train())
```

# Inheritance 

Note - Inheritance lets a class reuse and extend the behavior of another class. Nearly every model class in 
scikit-learn and every neural network layer in PyTorch is built using inheritance. 

```python
class BaseModel:
    def __init__(self,name):
        self.name = name

    def describe(self):
        return f"Model: {self.name}"

class ClassifierModel(BaseModel):          # inherits from BaseModel
    def __init__(self,name , num_classes):
        super().__init__(name)             # calls the parent's constructor
        self.num_classes = num_classes

m = ClassifierModel("LogReg",3)
print(m.describe())
print(m.num_classes)
```

Every custom PyTorch model is written as class MyModel(nn.Module): recognizing this pattern 
(inheriting from a base class, calling super().__init__()) is essential to reading any deep learning code.

 ##### note
super().__init__() ensures the parent class's setup logic still runs, forgetting it is a common bug 
when subclassing.
