## Lists

```python
nums = [4, 2, 9, 1] 
nums.append(7)        # add to end 
nums.sort()            # in-place sort 
nums[0]                # indexing -> 1 
nums[1:3]              # slicing -> [2, 4] 
nums[-1]               # last item 
len(nums) 
3 in nums              # membership test
```

```python
nums = [1, 2, 3] 
nums.extend([4, 5])   # [1,2,3,4,5] -- merge another list in 
nums.insert(0, 0)     # insert at a specific position -> [0,1,2,3,4,5] 
nums.remove(3)        # removes the VALUE 3 (first occurrence) 
nums.pop()            # removes & returns last item
nums.pop(0)           # removes & returns item at index 0 
nums.count(2)         # how many times 2 appears
nums.reverse()        # reverse in place
```

## *args & **kwargs

*args collects extra positional arguments into a tuple; **kwargs collects extra keyword arguments into a 
dict. You don't need to write these often, but ML library function signatures use them constantly.

```python
def train_model(*args, **kwargs):
    print(args)      # tuple of positional args
    print(kwargs)    # dictionary of keyword args


train_model(1, 2, lr=0.01, epochs=5) 
```

```python
# Unpacking when CALLING a function 
params = {"lr": 0.01, "epochs": 5} 
train_model(**params)   # spreads dict as keyword arguments
```

```python
# mistake - The names args and kwargs are convention, not syntax --- the * and ** are what matter, not the variable name. 
```

# Functions 
A reusable, named block of code that takes inputs (parameters) and optionally returns an output.

```python
def normalize(x, min_value=0, max_value=1):
    """Scale x between min_value and max_val."""
    return (x-min_value)/(max_value - min_value)

result = normalize(50,0,100)
print(result)

# inside functions, """ triple quotes strings """ , makes it a docString -- used for storing in documentation.
    
```

```python
def clean_text(text):
    text = text.lower()
    text = text.split()
    return text

clean_text("  HELLO World   ")
```

## List & Dictionary Comprehensions

A compact, Pythonic way to build a list or dictionary in one line instead of a multi-line loop

```python
squares = [x**2 for x in range(6)]
squares
```

```python
evens = [x%2==0 for x in range(10)]
evens
```

```python
evens = [x for x in range(10) if x%2==0]
evens
```

```python
label_map = {label : idx for idx, label in enumerate(['cat','dog','bird'])}
label_map
```

```python
# nested comprehension -- flatten a 2D list
matrix = [ [1,2],[3,4] ]
flat = [x for row in matrix for x in row]
flat
```

```python
names = ["asha", "rohan", "mira"] 
capitalized = [name.title() for name in names] 
names
```
